```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
# Parameters
num_particles = 6000  # Number of particles
mean_radius = 100     # Mean radius of the circle
std_dev = 12           # Standard deviation for the normal distribution
center_x, center_y = 150, 150  # Center of the circle
image_size = 300      # Size of the image (width, height)
small_circle_radius = mean_radius / 2  # Radius of the new solid circle

# Generate random angles for the particles (from 0 to 2π)
angles = np.random.uniform(0, 2 * np.pi, num_particles)

# Generate radii using a normal distribution with the given mean and standard deviation
radii = np.random.normal(mean_radius, std_dev, num_particles)

# Convert polar coordinates (r, theta) to Cartesian coordinates (x, y)
x_coords = center_x + radii * np.cos(angles)
y_coords = center_y + radii * np.sin(angles)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot particles as small circles
ax.scatter(x_coords, y_coords, s=2, color='blue', alpha=0.6, label="Diffusing particles")

# Plot the center point
ax.scatter(center_x, center_y, marker='x', color='red', s=25, label='Center')

# Add a label for the center point
ax.text(center_x - 12*2.1, center_y - 18, 'SNR Center', fontsize=12, color='red')

# Offset to move the starting points of the arrows slightly away from the center
offset = 50  # Adjust this value to move the arrow starting points further from the center
length_scale = 1.1

# Add arrows extending radially from the center with a small offset
ax.annotate('', xy=(center_x, center_y + mean_radius*length_scale), xytext=(center_x, center_y + offset),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('', xy=(center_x, center_y - mean_radius*length_scale), xytext=(center_x, center_y - offset),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('', xy=(center_x - mean_radius*length_scale, center_y), xytext=(center_x - offset, center_y),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('', xy=(center_x + mean_radius*length_scale, center_y), xytext=(center_x + offset, center_y),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Set the plot limits to frame the circle
ax.set_xlim(0, image_size)
ax.set_ylim(0, image_size)



# Hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

# Set equal aspect ratio
ax.set_aspect('equal')

plt.savefig("slides-diagrams/slide1.png")
plt.show()
```

```python
# Parameters
num_particles = 6000  # Number of particles
mean_radius = 100     # Mean radius of the circle
std_dev = 30           # Standard deviation for the normal distribution
center_x, center_y = 150, 150  # Center of the circle
image_size_x = 1000      # Size of the image (width, height)
image_size_y = 800      # Size of the image (width, height)
small_circle_radius = mean_radius / 2  # Radius of the new solid circle

# Generate random angles for the particles (from 0 to 2π)
angles = np.random.uniform(0, 2 * np.pi, num_particles)

# Generate radii using a normal distribution with the given mean and standard deviation
radii = np.random.normal(mean_radius*2, std_dev, num_particles)

# Convert polar coordinates (r, theta) to Cartesian coordinates (x, y)
x_coords = center_x + radii * np.cos(angles)
y_coords = center_y + radii * np.sin(angles)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot particles as small circles
ax.scatter(x_coords, y_coords, s=0.2, color='blue', alpha=0.6, label="Diffusing particles")

# Plot the center point
ax.scatter(center_x, center_y, marker='x', color='red', s=25, label='Center')

# Set the plot limits to frame the circles
ax.set_xlim(0, image_size_x)
ax.set_ylim(0, image_size_y)

# Plot a solid circle to the right of the original plot
cloud_circle_center_x = center_x + 2.4 * mean_radius  # Place it to the right of the original plot
cloud_circle_center_y = center_y + 2.4 * mean_radius # Keep the same y-coordinate

# Create the solid circle
solid_circle = Circle((cloud_circle_center_x, cloud_circle_center_y), small_circle_radius, color='green', alpha=0.7)

# Add a label for the Molecular cloud
ax.text(cloud_circle_center_x - 120, cloud_circle_center_y + 80, 'Spherical Molecular Cloud', fontsize=12, color='green')

# Add the solid circle to the plot
ax.add_patch(solid_circle)

R_scale_factor=1.1

# Plot earth
earth_center_x = abs(center_x - cloud_circle_center_x)*4
earth_center_y = abs(center_y - cloud_circle_center_y)*R_scale_factor
ax.scatter([earth_center_x], [earth_center_y])
ax.text(earth_center_x-80, earth_center_y - 40, "Earth", fontsize=10, color="black")

# Draw a straight line connecting the centers of the two circles
ax.plot([earth_center_x, cloud_circle_center_x], [earth_center_y, cloud_circle_center_y], color='black', linestyle='-', linewidth=1.3)
d_scale_factor = 1
ax.text(abs(cloud_circle_center_x - earth_center_x)*d_scale_factor, abs(cloud_circle_center_y - earth_center_y)*d_scale_factor, "d, distance to earth", fontsize=10, color="black")

# Plot dashed arrows extending radially from the surface of the solid circle
arrow_angles = np.linspace(0, 2 * np.pi, 8)  # 8 equally spaced angles for arrows
arrow_length = 90  # Length of the arrows

for angle in arrow_angles:
    # Calculate start and end points for each arrow
    start_x = cloud_circle_center_x + small_circle_radius * np.cos(angle)
    start_y = cloud_circle_center_y + small_circle_radius * np.sin(angle)
    end_x = start_x + arrow_length * np.cos(angle)
    end_y = start_y + arrow_length * np.sin(angle)
    
    # Add dashed arrows
    ax.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y),
                arrowprops=dict(arrowstyle='->', color='black', linestyle='--'))

ax.text(cloud_circle_center_x - 120, cloud_circle_center_y + 140, '(Gamma ray emission from cloud)', fontsize=10, color='black')

# Hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

# Set equal aspect ratio
ax.set_aspect('equal')

plt.savefig("slides-diagrams/slide3.png")
plt.show()
```

```python
# Parameters
num_particles = 6000  # Number of particles
mean_radius = 100     # Mean radius of the circle
std_dev = 30           # Standard deviation for the normal distribution
center_x, center_y = 150, 150  # Center of the circle
image_size_x = 600      # Size of the image (width, height)
image_size_y = 600      # Size of the image (width, height)
small_circle_radius = mean_radius / 2  # Radius of the new solid circle

# Generate random angles for the particles (from 0 to 2π)
angles = np.random.uniform(0, 2 * np.pi, num_particles)

# Generate radii using a normal distribution with the given mean and standard deviation
radii = np.random.normal(mean_radius*2, std_dev, num_particles)

# Convert polar coordinates (r, theta) to Cartesian coordinates (x, y)
x_coords = center_x + radii * np.cos(angles)
y_coords = center_y + radii * np.sin(angles)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Plot particles as small circles
ax.scatter(x_coords, y_coords, s=0.2, color='blue', alpha=0.6, label="Diffusing particles")

# Plot the center point
ax.scatter(center_x, center_y, marker='x', color='red', s=25, label='Center')

# Set the plot limits to frame the circles
ax.set_xlim(0, image_size_x)
ax.set_ylim(0, image_size_y)

# Plot a solid circle to the right of the original plot
cloud_circle_center_x = center_x + 2.4 * mean_radius  # Place it to the right of the original plot
cloud_circle_center_y = center_y + 2.4 * mean_radius # Keep the same y-coordinate

# Create the solid circle
solid_circle = Circle((cloud_circle_center_x, cloud_circle_center_y), small_circle_radius, color='green', alpha=0.7)

# Add the solid circle to the plot
ax.add_patch(solid_circle)

R_scale_factor=1.1

# Plot dashed arrows extending radially from the surface of the solid circle
arrow_angles = np.linspace(0, 2 * np.pi, 8)  # 8 equally spaced angles for arrows

arrow_length = 90  # Length of the arrows
def getArrowLength(angle):
    # return arrow_length
    return np.abs(np.sin(angle/2 + 2.6)) * arrow_length

for angle in arrow_angles:
    # Calculate start and end points for each arrow
    start_x = cloud_circle_center_x + small_circle_radius * np.cos(angle)
    start_y = cloud_circle_center_y + small_circle_radius * np.sin(angle)
    end_x = start_x + getArrowLength(angle) * np.cos(angle)
    end_y = start_y + getArrowLength(angle) * np.sin(angle)
    
    # Add dashed arrows
    ax.annotate('', xy=(end_x, end_y), xytext=(start_x, start_y),
                arrowprops=dict(arrowstyle='->', color='black', linestyle='--'))


# Hide axes
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

# Set equal aspect ratio
ax.set_aspect('equal')

plt.savefig("slides-diagrams/differential-emission.png")
plt.show()
```
